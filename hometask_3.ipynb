```python
#1
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt('http://scipy-lectures.org/_downloads/populations.txt')
data = data.astype(float)
datah = data[:,1]
datay = data[:,0]
datal = data[:,2]
datac = data[:,3]
#print(datax)
plt.plot(datay,datah)
plt.title('hare')
plt.xlabel('year')
plt.ylabel('hare')
plt.show()
plt.plot(datay,datac)
plt.title('carrot')
plt.xlabel('year')
plt.ylabel('carrot')
plt.show()
plt.plot(datay,datah)
plt.title('lynx')
plt.xlabel('year')
plt.ylabel('lynx')
plt.show()
```

```python
#2
fig, ax = plt.subplots()

ax.bar(data[:,0],1)
ax.bar(data[:,0],data[:,2]/(data[:,2]+data[:,1]))
plt.legend(('hare', 'lynx'), loc='upper right')
plt.show()
```

```python
#3
plt.figure(figsize=(17,10))
plt.subplot(3, 3, 1)
plt.hist(data[:,1], bins='auto')  # arguments are passed to np.histogram
plt.title("hare")
plt.axvline(x=np.mean(data[:,1]),color = 'r')
plt.subplot(3, 3, 2)
plt.hist(data[:,2], bins='auto')  # arguments are passed to np.histogram
plt.title("lynx")
plt.axvline(x=np.mean(data[:,2]),color = 'r')
plt.subplot(3, 3, 3)
plt.hist(data[:,3], bins='auto')  # arguments are passed to np.histogram
plt.title("carrot")
plt.axvline(x=np.mean(data[:,3]),color = 'r')
plt.show()
```

```python
#4
a = np.array([(np.argmax(data[i, 1:]), i) for i in range(len(data[:,0]))])
a = np.transpose(a)
ah = a[1][a[0] == 0]
al = a[1][a[0] == 1]
ac= a[1][a[0] == 2]
plt.plot(data[:,0], data[:,1])
plt.scatter(data[ah,0], data[ah,1], c='r')
plt.title('Hares')
plt.legend(('All', 'Highest of all'), loc='upper right')
plt.show()
plt.plot(data[:,0], data[:,2])
plt.scatter(data[al,0], data[al,2], c='g')
plt.title('lynx')
plt.legend(('All', 'Highest of all'), loc='upper right')
plt.show()
plt.plot(data[:,0], data[:,3])
plt.scatter(data[ac,0], data[ac,3], c='y')
plt.title('carrot')
plt.legend(('All', 'Highest of all'), loc='upper right')
plt.show()
```

```python
#5
mask = datal > datah
x = np.zeros(len(mask))
x[mask] = 1
x1 = np.array([int(x[i]) * "L"+ (1 -int(x[i])) * "H" for i in range(len(x))])
a = [sum(x) / len(x), (len(x) - sum(x) )/len(x)] 
print(x1)
plt.pie(a, labels = ['lynx', 'hare'], autopct='%1.0f%%', radius = 1 )
plt.show()
```

```python
#6
x = data[:,0]
yh1 = np.ma.masked_where(data[:,1] < 50000, data[:,1])
yh2 = data[:, 1]
yl1 = np.ma.masked_where(data[:,2] < 50000, data[:,2])
yl2 = data[:, 2]
yc1 = np.ma.masked_where(data[:,3] < 50000, data[:,3])
yc2 = data[:, 3]

l0 = plt.plot(x, yh1)
l1 = plt.plot(x, yh2)
plt.setp(l1, c='r')
plt.setp(l0, linewidth=4, c='black')
plt.scatter(x, yh1, c='black')
plt.legend(('>50000', 'Every'), loc ='upper right')
plt.title('Hares')
plt.show()

l0 = plt.plot(x, yl1)
l1 = plt.plot(x, yl2)
plt.setp(l1, c='g')
plt.setp(l0, linewidth=4, c='black')
plt.scatter(x, yl1, c='black')
plt.legend(('>50000', 'Every'), loc ='upper right')
plt.title('lynx')
plt.show()

l0 = plt.plot(x, yc1)
l1 = plt.plot(x, yc2)
plt.setp(l1, c='b')
plt.setp(l0, linewidth=4, c='black')
plt.scatter(x, yc1, c='black')
plt.legend(('>50000', 'Every'), loc ='upper right')
plt.title('carrot')
plt.show()
```

```python
#7
yl = np.argsort(data[:, 1])[:2]
yh = np.argsort(data[:, 2])[:2]
yc = np.argsort(data[:, 3])[:2]

plt.plot(data[:,0], data[:,1])
plt.scatter(data[yl,0], data[yl,1], c='r')
plt.title('Hares')
plt.legend(('All', 'Lowest'), loc='upper right')
plt.show()

plt.plot(data[:,0], data[:,2])
plt.scatter(data[yh,0], data[yh,2], c='r')
plt.title('lynx')
plt.legend(('All', 'Lowest'), loc='upper right')
plt.show()

plt.plot(data[:,0], data[:,3])
plt.scatter(data[yc,0], data[yc,3], c='r')
plt.title('carrot')
plt.legend(('All', 'Lowest'), loc='upper right')
plt.show()
```

```python
#8
a1 = data[:,1]
a11 = np.delete(a1,0)
a11 = np.append(a11,np.mean(a1))
a = a11-a1
print("corr = ",np.corrcoef(a, data[:,2])[0,1])
plt.plot(data[:,0], data[:,2])
plt.plot(data[:,0], a)
plt.legend(('change in hare population ', 'lynx'), loc='upper right')
plt.show()
```

```python
#9
h = np.argsort(data[:, 1])
plt.figure(figsize=(17,10))
plt.subplot(2, 2, 1)
plt.plot(data[h, 1], data[h, 2], c='r')
plt.title('hare vs lynxe')
plt.xlabel('hare')
plt.ylabel('lynxe')

plt.subplot(2, 2, 2)
plt.plot(data[h, 1], data[h, 3], c='r')
plt.title('hare vs carrot')
plt.xlabel('hare')
plt.ylabel('carrot')
plt.show()
```
